# Starting EDA with high level statsÂ¶

```python
country_counts = population_vert.groupby(['CountryCode']).agg({'value':sum})
country_level = pd.DataFrame(country_counts.reset_index())
country_level = country_level.sort_values(by=['value'],ascending=False)
country_level
```

```python
genpop = pd.read_excel('QES_genpop.xlsx')
genpop = genpop.sort_values(by=['Population_2012'], ascending=False)
genpop
```

```python
genpop['sample'] = genpop['CountryCode'].map(country_level.set_index('CountryCode')['value'])
genpop['respondents_country']= genpop['sample']/genpop['Population_2012']*100
genpop = genpop.sort_values(by=['respondents_country'], ascending=False)
genpop.to_csv('genpop.csv')
genpop
```

```python
query = '''SELECT question_code, question_label,answer,SUM(response)
FROM res_all
GROUP BY question_code,question_label, answer;'''

all_answers = pd.read_sql_query(query, engine)
all_answers.head()
```

```python
all_answers['percent_all'] = round(all_answers['SUM(response)']/93079*100,2)
all_answers.head()
```

```python
query = '''SELECT CountryCode, question_code, question_label,answer,SUM(response)
FROM res_all
GROUP BY question_code,question_label, answer,CountryCode;'''

g2_input = pd.read_sql_query(query, engine)
g2_input.head()
```

```python
g2 = g2_input[g2_input['question_code'].isin(['g2_a','g2_b','g2_c','g2_d'])]
g2 = g2.groupby(['answer','CountryCode']).agg({"SUM(response)":sum})
g2 = pd.DataFrame(g2)
g2.to_csv('g2.csv')
g2
```

```python
g2_pivot = g2.pivot_table(index= ['CountryCode'],columns=['answer'])
g2_pivot.to_csv('g2_pivot.csv')
g2_pivot
```
